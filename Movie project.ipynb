# Classify movies based on ratings(excellent,good and average)

```python
#create our own function
def rating(rating):
        if rating>=7.0:
            return "Excellent"
        elif rating>=6.0:
            return "Good"
        else:
            return "Average"
```

```python
#lets create new column for this categories
df['Rating_category']=df['Rating'].apply(rating)
```

# Show how many films of each genre were made

```python
#first we split the genres using comas and make an empty list
list1=[]
for value in df['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
#we going to add this 2list to 1list
one_d=[]
for item in list1:
        for item1 in item:
            one_d.append(item1)
```

```python
#lets check our 1list
one_d
```

```python
#lets find unique values
uni_list=[]
for item in one_d:
        if item not in uni_list:
            uni_list.append(item)
```

```python
uni_list
```

```python
#lets find how many films of each genre were made
from collections import Counter
```

```python
#lets visualize it
# Convert the flattened list into a DataFrame
df_one_d = pd.DataFrame(one_d, columns=['Genre'])

# Count the occurrences of each category
genre_counts = df_one_d['Genre'].value_counts().reset_index()

# Rename the columns for clarity
genre_counts.columns = ['Genre', 'Count']

# Create a bar plot to visualize the number of occurrences per category
sns.barplot(x='Count', y='Genre', data=genre_counts, orient='h')
plt.title("Number of films of each genre made")
plt.xlabel("Number of Occurrences")
plt.ylabel("Genre")
plt.show()
```
